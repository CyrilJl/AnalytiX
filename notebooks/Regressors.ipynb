```python
import pandas as pd
from analytix import XGBRegressorWrapper
from sklearn.datasets import load_diabetes
```

```python
data = load_diabetes()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

```python
%%time
model_xgb = XGBRegressorWrapper().fit(X, y)
```

```python
model_xgb.beeswarm(X)
```

```python
model_xgb.dependence(X=X, feature='bmi')
```
